## Subplot
Me permite generar mÃ¡s visualizaciones de acuerdo a mis necesidades

````python
plt.subplot(1,2,1) #tres parametros, numero de filas, n de columnas, index que voy a manejar entre filas y columnas
plt.plot(x,y,'g--') #que quiero en ese grafico que defini
plt.subplot(1,2,2) #en el segundo grafico en index es el 2
plt.hist(y) #definicion del segundo grafico
plt.show 
````

````python
plt.subplot(1,2,1) #tres parametros, numero de filas, n de columnas del arreglo grafico y el tercer valor es el index que voy a manejar entre filas y columnas
#es decir ese tercer parÃ¡metro indica donde estarÃ¡ este primer grÃ¡fico, en este caso indice 1
plt.plot(x,y,'g--') #que quiero en ese grafico que defini
plt.plot(y,x,'b:') #Todo lo que ponga bajo la definiciÃ³n del arreglo grafico saldrÃ¡ en ese indice (1)
plt.subplot(1,2,2) #en el segundo grafico en index 2
plt.pie(y) #definicion del segundo grafico que saldrÃ¡ en la posiciÃ³n (2)
plt.show
````

## MÃ©todo Orientado a Objetos

MÃ¡s constumizable, mÃ¡s a amigable a mÃ¡s diagramas, requiere mÃ¡s cÃ³digo

````python
from matplotlib import axes
#Primero creo el linezo donde vaa a estar mi grÃ¡fica
fig = plt.figure()

#Nuestro grÃ¡ficos serÃ¡n los "axes", a los cuales le agregamos
axes=fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
````

````python
#La ventaja viene cuando quiero hacer algo mucho mÃ¡s especifico
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.9])
#los dos primero parÃ¡metros hacen referencia a la posiciÃ³n de la figura en el lienzo
#los siguientes 2, definen el tamaÃ±o del axes (grÃ¡fico) en el lienzo (horizontal,vertical)
axes2=fig.add_axes([0.99,-0.65,0.4,0.3]) #para hacer un segundo grÃ¡fico

axes.plot(x,y,'b')
axes2.plot(y,x,'g')
axes2.set_facecolor('lightgray') #asÃ­ puedo modificar el color de fondo de ese axes
#Esta es una ventaja de usar mÃ©todo orientado a objeto, cada objeto es distinto y lo puedo modificar por separado

fig.show()
````

````python
np.random.seed(19680801)

# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)
'''
mu: Representa la media del conjunto de datos, en este caso 100.
sigma: Representa la desviaciÃ³n estÃ¡ndar, que indica la variabilidad de los datos respecto a la media, en este caso 15.
x: Creamos un array de tamaÃ±o 437 utilizando la funciÃ³n np.random.randn.
Esta funciÃ³n genera nÃºmeros aleatorios siguiendo una distribuciÃ³n normal (de Gauss) con media mu y desviaciÃ³n estÃ¡ndar sigma.
'''

num_bins = 50 #La variable num_bins define el nÃºmero de barras que tendrÃ¡ el histograma. En este caso, se usarÃ¡n 50 barras para representar la distribuciÃ³n de los datos.



fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

'''
fig, ax: Se crea una nueva figura (fig) y un objeto de ejes (ax) para el histograma.
ax.hist: Se crea el histograma a partir de los datos en x, usando num_bins barras y la opciÃ³n density=True para representar la densidad de probabilidad en lugar de conteos simples.
'''


# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
'''
En estas lÃ­neas se aÃ±ade una lÃ­nea que representa la distribuciÃ³n normal (de Gauss) esperada con los valores de mu y sigma definidos anteriormente. 
La fÃ³rmula para calcular la curva se basa en la funciÃ³n de densidad de probabilidad de la distribuciÃ³n normal.

'''


ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

'''
e establecen las etiquetas para los ejes X (Smarts - Inteligencia) e Y (Probability density - Densidad de probabilidad). 
AdemÃ¡s, se agrega un tÃ­tulo al histograma que muestra la media (mu) y la desviaciÃ³n estÃ¡ndar (sigma) de la distribuciÃ³n.
'''

fig.tight_layout() # Esta lÃ­nea utiliza la funciÃ³n fig.tight_layout para ajustar el espacio entre el histograma y las etiquetas, evitando que se superpongan.
plt.show()
````
