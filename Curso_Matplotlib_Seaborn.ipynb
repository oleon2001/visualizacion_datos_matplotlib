## Subplot
Me permite generar más visualizaciones de acuerdo a mis necesidades

````python
plt.subplot(1,2,1) #tres parametros, numero de filas, n de columnas, index que voy a manejar entre filas y columnas
plt.plot(x,y,'g--') #que quiero en ese grafico que defini
plt.subplot(1,2,2) #en el segundo grafico en index es el 2
plt.hist(y) #definicion del segundo grafico
plt.show 
````

````python
plt.subplot(1,2,1) #tres parametros, numero de filas, n de columnas del arreglo grafico y el tercer valor es el index que voy a manejar entre filas y columnas
#es decir ese tercer parámetro indica donde estará este primer gráfico, en este caso indice 1
plt.plot(x,y,'g--') #que quiero en ese grafico que defini
plt.plot(y,x,'b:') #Todo lo que ponga bajo la definición del arreglo grafico saldrá en ese indice (1)
plt.subplot(1,2,2) #en el segundo grafico en index 2
plt.pie(y) #definicion del segundo grafico que saldrá en la posición (2)
plt.show
````

````python
#Modificando disposición de los graficos, ojo con el manejo de los index
plt.subplot(2,1,1)
plt.plot(x,y,'g--')
plt.plot(x,x,'g--')
plt.plot(y,x,'r:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
````

## Método Orientado a Objetos

Más constumizable, más a amigable a más diagramas, requiere más código

````python
from matplotlib import axes
#Primero creo el linezo donde vaa a estar mi gráfica
fig = plt.figure()

#Nuestro gráficos serán los "axes", a los cuales le agregamos
axes=fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
````

````python
#La ventaja viene cuando quiero hacer algo mucho más especifico
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.9])
#los dos primero parámetros hacen referencia a la posición de la figura en el lienzo
#los siguientes 2, definen el tamaño del axes (gráfico) en el lienzo (horizontal,vertical)
axes2=fig.add_axes([0.99,-0.65,0.4,0.3]) #para hacer un segundo gráfico

axes.plot(x,y,'b')
axes2.plot(y,x,'g')
axes2.set_facecolor('lightgray') #así puedo modificar el color de fondo de ese axes
#Esta es una ventaja de usar método orientado a objeto, cada objeto es distinto y lo puedo modificar por separado

fig.show()
````

````python
np.random.seed(19680801)

# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)
'''
mu: Representa la media del conjunto de datos, en este caso 100.
sigma: Representa la desviación estándar, que indica la variabilidad de los datos respecto a la media, en este caso 15.
x: Creamos un array de tamaño 437 utilizando la función np.random.randn.
Esta función genera números aleatorios siguiendo una distribución normal (de Gauss) con media mu y desviación estándar sigma.
'''

num_bins = 50 #La variable num_bins define el número de barras que tendrá el histograma. En este caso, se usarán 50 barras para representar la distribución de los datos.



fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

'''
fig, ax: Se crea una nueva figura (fig) y un objeto de ejes (ax) para el histograma.
ax.hist: Se crea el histograma a partir de los datos en x, usando num_bins barras y la opción density=True para representar la densidad de probabilidad en lugar de conteos simples.
'''


# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
'''
En estas líneas se añade una línea que representa la distribución normal (de Gauss) esperada con los valores de mu y sigma definidos anteriormente. 
La fórmula para calcular la curva se basa en la función de densidad de probabilidad de la distribución normal.

'''


ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

'''
e establecen las etiquetas para los ejes X (Smarts - Inteligencia) e Y (Probability density - Densidad de probabilidad). 
Además, se agrega un título al histograma que muestra la media (mu) y la desviación estándar (sigma) de la distribución.
'''

fig.tight_layout() # Esta línea utiliza la función fig.tight_layout para ajustar el espacio entre el histograma y las etiquetas, evitando que se superpongan.
plt.show()
````
